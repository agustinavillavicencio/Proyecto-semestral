```python
import pandas as pd

df = pd.read_csv('/content/BASE 3 OFICIAL.csv')

df = pd.read_csv('/content/BASE 3 OFICIAL.csv', sep=';')

pd.set_option('display.max_rows', None)
df
```

```python
import altair as alt
```

```python
color_scale = alt.Scale(
    domain=df['Artista'].unique().tolist(),
    range=['#FF00C8', '#C6FF1A', '#C9A2E6']
)

chart = alt.Chart(df).mark_line(point=True).encode(
    x=alt.X('Numero de era:O', title=None),
    y=alt.Y('Promedio de transformación:Q',
            title='Promedio de transformación',
            scale=alt.Scale(domainMin=0.8, domainMax=2.2)),
    color=alt.Color('Artista:N', scale=color_scale, legend=None),
    tooltip=['Artista', 'Año', 'Album', 'Promedio de transformación']
).properties(
    width=100
).facet(
    column=alt.Column('Artista:N',
                      title="Número de Era (álbum)",
                      header=alt.Header(titleOrient="bottom", labelOrient="bottom"))
).properties(
    title=alt.Title(
        'Transformación por Era (álbum)',
        align='center',
        anchor='middle'
    )
)

chart
```
