```python
import pandas as pd
import altair as alt

# Cargar datos con codificación compatible
df = pd.read_excel("/content/popstars_codificados_enteros_corregido.xlsx")

# Preparar datos
plot_cols = [col for col in df.columns if col != 'Artistas femeninas' and 'ranking' not in col.lower()]
df_melted = df.melt(id_vars='Artistas femeninas', value_vars=plot_cols,
                    var_name='Característica', value_name='Valor')

# Gráfico de áreas traslapadas
overlap_area = alt.Chart(df_melted).mark_area(
    interpolate='monotone',
    opacity=0.25
).encode(
    x=alt.X('Característica:N', sort=None, axis=alt.Axis(labelAngle=45)),
    y=alt.Y('Valor:Q', stack=None),
    color=alt.Color('Artistas femeninas:N'),
    tooltip=['Artistas femeninas:N', 'Característica:N', 'Valor:Q']
).properties(
    title='Soundwave traslapado: Comparación de perfiles',
    width=800,
    height=400
)

overlap_area
```

# un poco más marcado con las líneas

```python
# Capa de área transparente
area_layer = alt.Chart(df_melted).mark_area(
    interpolate='monotone',
    opacity=0.3
).encode(
    x=alt.X('Característica:N', sort=None, axis=alt.Axis(labelAngle=45)),
    y=alt.Y('Valor:Q', stack=None),
    color=alt.Color('Artistas femeninas:N')
)

# Capa de línea encima
line_layer = alt.Chart(df_melted).mark_line(
    interpolate='monotone',
    opacity=0.6,
    strokeWidth=1.5
).encode(
    x='Característica:N',
    y=alt.Y('Valor:Q', stack=None),
    color=alt.Color('Artistas femeninas:N'),
    tooltip=['Artistas femeninas:N', 'Característica:N', 'Valor:Q']
)

# Combinar ambas capas
wave_overlay = (area_layer + line_layer).properties(
    title='Soundwave superpuesto: Área + línea (comparación de perfiles)',
    width=800,
    height=400
)

wave_overlay
```

```python
wave_overlay.save('Global visualización 1.html')
```
