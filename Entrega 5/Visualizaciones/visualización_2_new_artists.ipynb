```python
import pandas as pd

df = pd.read_csv("/content/Base_codificada_artistas_nuevas.csv", sep=';')
df.head(12)
```

```python
import pandas as pd
import altair as alt
```

```python
df_melted = df.melt(
    id_vars=['Artista'],
    value_vars=[
        'Ojos',
        'altura',
        'contextura corporal',
        'simetría facial',
        'labios',
        'perfil facial',
        'cabello',
        'color de piel',
        'estética musical dinámica/marcada',
        'presencia escénica (shows con narrativa, baile, etc.)',
        'edad de peak en éxito',
        'cantidad de Interacción con el público',
        'tipo de interacción con el público',
        'Vestuario apariciones públicas',
        'Estilo interpretativo',
        'Narrativa personal',
        'impacto cultural',
        'presencia digital y multiplataforma'
    ],
    var_name='Característica',
    value_name='Valor'
)
```

```python
area_facet = alt.Chart(df_melted).mark_area(
    interpolate='monotone',
    opacity=0.4,
    stroke='black'
).encode(
    x=alt.X('Característica:N', axis=alt.Axis(labelAngle=45)),
    y=alt.Y('Valor:Q', scale=alt.Scale(domain=[
        df_melted['Valor'].min(), df_melted['Valor'].max()
    ])),
    tooltip=['Característica:N', 'Valor:Q'],
    color=alt.Color('Artista:N',
        scale=alt.Scale(
            range=['#C6FF1A', '#D5FB57', '#EBB6D6', '#FF00C8', '#C9A2E6', '#f26fff']
        ),
        legend=None
    )
).facet(
    column=alt.Column('Artista:N', header=alt.Header(labelAngle=45))
).properties(
    title='Comparación de características por artista pop (visualización 2)',
).configure_view(
    stroke=None
)

area_facet
```

```python
area_facet.save('visualizacion 2 ARTISTAS NUEVAS.html')
```
