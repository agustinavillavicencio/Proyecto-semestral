```python
import altair as alt
```

```python

file_path = "/content/rasgos de una pop star.csv"
df = pd.read_csv(file_path, sep=';')

df = df.drop(columns=['Posición en ranking Billboard'])

df_long = df.melt(id_vars='Artistas femeninas', var_name='Rasgo', value_name='Valor')

artistas = df_long['Artistas femeninas'].unique()
rasgos = df_long['Rasgo'].unique()

valores_por_rasgo = df_long.groupby('Rasgo')['Valor'].unique().to_dict()

combinaciones = []
for artista in artistas:
    for rasgo in rasgos:
        for valor in valores_por_rasgo[rasgo]:
            combinaciones.append((artista, rasgo, valor))

df_combinaciones = pd.DataFrame(combinaciones, columns=['Artistas femeninas', 'Rasgo', 'Valor'])

df_merge = pd.merge(df_combinaciones, df_long, on=['Artistas femeninas', 'Rasgo', 'Valor'], how='left', indicator=True)

df_merge['Puntaje'] = df_merge['_merge'].apply(lambda x: 10 if x == 'both' else 0)

df_merge['Rasgo_Valor'] = df_merge['Rasgo'] + ': ' + df_merge['Valor']

chart = alt.Chart(df_merge).mark_area(interpolate='step').encode(
    x=alt.X('Rasgo_Valor:N', sort=None, title='Rasgos y valores'),
    y=alt.Y('Puntaje:Q'),
    color=alt.Color('Artistas femeninas:N', scale=alt.Scale(scheme='pinkyellowgreen')),
    row=alt.Row('Artistas femeninas:N', title=None),
    tooltip=['Artistas femeninas', 'Rasgo', 'Valor', 'Puntaje']
).properties(
    height=50,
    width=900,
    title='Perfil de rasgos por artista (0 = no lo tiene, 10 = sí lo tiene)'
).configure_axis(
    labelFontSize=10,
    titleFontSize=12,
    labelAngle=45
)

chart
```
